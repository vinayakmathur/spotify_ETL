```python
songs = []
artists = []
played_at_list=[]
timestramps=[]
for song in j['items']:
    songs.append(song['track']['name'])
    artists.append(song["track"]["album"]["artists"][0]["name"])
    played_at_list.append(song["played_at"])
    timestramps.append(song["played_at"][0:10])
```
